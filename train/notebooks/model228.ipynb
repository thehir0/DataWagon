```python
x_train = pd.get_dummies(x_train, columns=['kuzov', 'tormoz', 'tipvozd', 'tippogl'])
x_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train[x_train.columns] = scaler.fit_transform(x_train[x_train.columns])
x_train.head()
```

```python
x_train = x_train.fillna(x_train.mean())
print(x_train.shape[1])
x_train.head()
```

```python
# save x_train and y_train
x_train.to_csv('interim/x_train.csv', index=False)
y_train.to_csv('interim/y_train.csv', index=False)
```

```python
# train decision tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

month_tree = DecisionTreeRegressor(random_state=42)
tree_params = {'max_depth': range(1, 11), 'max_features': range(1, 11)}
month_tree_grid = GridSearchCV(month_tree, tree_params, cv=5, n_jobs=-1, verbose=True)
month_tree_grid.fit(x_train, y_train_month)

day_tree = DecisionTreeRegressor(random_state=42)
day_tree_grid = GridSearchCV(day_tree, tree_params, cv=5, n_jobs=-1, verbose=True)
day_tree_grid.fit(x_train, y_train_day)

print('Month', month_tree_grid.best_params_, month_tree_grid.best_score_)
print('Day', day_tree_grid.best_params_, day_tree_grid.best_score_)
```
