```python
import pandas as pd
import sys
sys.path.append('..')
from src.timeconverter import date_to_timestamp
```

```python
train1 = pd.read_csv('./target/y_train.csv')
train2 = pd.read_csv('../train2/target/y_test.csv')
train = pd.concat([train1, train2])
```

```python
wag_param = pd.read_csv('../preprocessed/wag_param.csv', index_col = 0)
kti = pd.read_csv('../preprocessed/kti.csv', index_col = 0)
tr_rem = pd.read_csv('../preprocessed/tr_rem_preprocessed.csv')
dislok = pd.read_csv('../preprocessed/dislok_and_freight.csv')

train = pd.read_csv('./target/y_train.csv')
train['month'] = train['month'].apply(date_to_timestamp)
```

```python
df = pd.merge(train, kti, how='left', on='wagnum')
df
```

```python
df['operation_date_dttm'].fillna(0, inplace=True)
```

```python
df.isna().sum()
```

```python
# Find rows where 'month' is greater than 'operation_date_dttm'
filtered_df = df[df['month'] >= df['operation_date_dttm']]
# Filter to keep rows where 'operation_date_dttm' is the maximum within 'wagnum' groups
filtered_df = filtered_df[filtered_df.groupby(['wagnum', 'month'])['operation_date_dttm'].transform('max') == filtered_df['operation_date_dttm']]
filtered_df
```

```python
filtered_df = pd.merge(filtered_df, wag_param, how='left', on='wagnum')
filtered_df = pd.merge(filtered_df, tr_rem, how='left', on='wagnum')
filtered_df = pd.merge(filtered_df, dislok, how='left', on='wagnum')
filtered_df
```

```python
def split(data):
    y1 = data['target_month']
    y2 = data['target_day']

    X = data.drop(['target_month', 'target_day'], axis=1)
    return X, y1, y2
```

```python
X_train, y1_train, y2_train = split(filtered_df)
X_train
```

```python
from catboost import CatBoostClassifier

cat_features = ['model']
# Initializing CatBoost Classifier
model1 = CatBoostClassifier(iterations=5000, loss_function='Logloss', random_seed=42, cat_features=cat_features, eval_metric='F1', nan_mode='Min')
model2 = CatBoostClassifier(iterations=5000, loss_function='Logloss', random_seed=42, cat_features=cat_features, eval_metric='F1', nan_mode='Min')

# Training the model
model1.fit(X_train, y1_train, eval_set=(X_train, y1_train))
model2.fit(X_train, y2_train, eval_set=(X_train, y2_train))
# # Model evaluation (if needed)
# metrics = model.get_evals_result()

# Feature importance
feature_importance1 = model1.get_feature_importance()
feature_importance2 = model1.get_feature_importance()

print(feature_importance1)
print(feature_importance2)
```

```python
y1_predicted = model1.predict(X_train)
y2_predicted = model2.predict(X_train)
```

```python
from sklearn.metrics import f1_score
score_month = f1_score(y1_predicted, y1_train)
score_day = f1_score(y2_predicted, y2_train)
score = 0.5 * score_month + 0.5 * score_day
score
```
